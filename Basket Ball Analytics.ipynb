# When you use Numpy arrays, all elements of your array (or matrix in this case) should have the same datatype. As your lists are not formed by the same data types, Numpy's best guess was they are lists of different elements, and to "warn" you of that, print the type of the element (list).

# It will occur the same if you have different size lists of numbers, for instance.

# But if you define 3 lists of just one data type and the same number of elements, Numpy will create a matrix (multidimensional array).

```python
r1 = ["I", 'am', 'happy']
r2 = ['What', 'a', 'day']    # Here there are equal number of elements in each lists although the datatypes are different.
r3 = [1, 2, 3]
```

```python
np.array([r1, r2, r3])
```

```python
print(np.array([r1, r2, r3]))
```

### Function

```python
def plot(req_matrix, playerList = Players):    # By default, if we don't pass any player's name, every player's details will be displayed.
    
    colors = {"KobeBryant":'Black',"JoeJohnson":'Blue',"LeBronJames":'Green',"CarmeloAnthony":'Red',"DwightHoward":'Yellow',"ChrisBosh":'Magenta',"ChrisPaul":'Indigo',"KevinDurant":'Violet',"DerrickRose":'Orange',"DwayneWade":'Brown'}
    markers = {"KobeBryant":'s',"JoeJohnson":'o',"LeBronJames":'v',"CarmeloAnthony":'*',"DwightHoward":'x',"ChrisBosh":'D',"ChrisPaul":'P',"KevinDurant":'|',"DerrickRose":'p',"DwayneWade":'X'}
    for name in playerList:
        plt.plot(req_matrix[Pdict[name]], c = colors[name], ls = '--', marker = markers[name], ms = 7, label = name)
        
    plt.legend(loc = 'upper left', bbox_to_anchor = (1, 1))
    plt.xticks(list(range(0, 10)), Seasons)
    plt.show()
```
